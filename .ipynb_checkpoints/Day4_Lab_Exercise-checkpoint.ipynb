```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import folium
```

```python
# importing data 1
primary = pd.read_csv("./education_analytics/Enrollment Master Data_2015_E.csv", encoding="latin-1")
primary.info()
secondary = pd.read_csv("./education_analytics/Enrollment Master Data_2015_S.csv", encoding="latin-1")
secondary.info()
schools = pd.read_csv("./education_analytics/Masterlist of Schools.csv", encoding="latin-1")
schools.info()
school_locs = pd.read_csv("./education_analytics/Schools Location Data.csv", encoding="latin-1")
school_locs.info()
```

```python
mooe = pd.read_csv(".\education_analytics\MOOE data.csv", encoding="latin-1")
mooe.info()
rooms = pd.read_csv(".\education_analytics\Rooms data.csv", encoding="latin-1")
rooms.info()
teachers = pd.read_csv(".\education_analytics\Teachers data.csv", encoding="latin-1")
teachers.info()
```

```python
#cleaning mooe data
mooe['school.mooe'] = mooe['school.mooe'].str.replace(',','').astype(float)
mooe['school.mooe']
#teachers.loc[teachers['ï»¿school.id'] == "320305"]
schools[schools['ï»¿school.id'] == "320305"].nunique()
#teachers.drop(index = 320305)
```

```python
#combining mooe, teachers, and rooms datasets
mt = pd.merge(mooe, teachers, left_on="ï»¿school.id", right_on="ï»¿school.id", how="inner")
mt.info()

mtr = pd.merge(mt, rooms, left_on="ï»¿school.id", right_on="ï»¿School ID", how="inner")
mtr.info()
```

```python
#creating master dataset
schools_master = pd.merge(schools, mtr, left_on="ï»¿school.id", right_on="ï»¿school.id", how="inner")
schools_master.info()
```

```python
#new column for mooe per student/teacher/school ratio

schools_master['enrol_mooe'] = schools_master['school.mooe'] / schools_master['school.enrollment']
schools_master['total_teachers'] = schools_master['teachers.instructor'] + schools_master['teachers.mobile'] + schools_master['teachers.regular'] + schools_master['teachers.sped']
schools_master['teacher_mooe'] = schools_master['school.mooe'] / schools_master['total_teachers']
schools_master['total_rooms'] = schools_master['rooms.standard.academic'] + schools_master['rooms.standard.unused'] + schools_master['rooms.nonstandard.academic'] + schools_master['rooms.nonstandard.unused']
schools_master['room_mooe'] = schools_master['school.mooe'] / schools_master['total_rooms']

#replace roman numerals in region names
schools_master['school.region'].replace("Region I", value="Region 01",inplace=True)
schools_master['school.region'].replace("Region II", value="Region 02",inplace=True)
schools_master['school.region'].replace("Region III", value="Region 03",inplace=True)
schools_master['school.region'].replace("Region IV-A", value="Region 04A",inplace=True)
schools_master['school.region'].replace("Region IV-B", value="Region 04B",inplace=True)
schools_master['school.region'].replace("Region V", value="Region 05",inplace=True)
schools_master['school.region'].replace("Region VI", value="Region 06",inplace=True)
schools_master['school.region'].replace("Region VII", value="Region 07",inplace=True)
schools_master['school.region'].replace("Region VIII", value="Region 08",inplace=True)
schools_master['school.region'].replace("Region IX", value="Region 09",inplace=True)
schools_master['school.region'].replace("Region X", value="Region 10",inplace=True)
schools_master['school.region'].replace("Region XI", value="Region 11",inplace=True)
schools_master['school.region'].replace("Region XII", value="Region 12",inplace=True)
schools_master['school.region'].unique()
```

```python
#creating dataset for all DepEd and LGU managed schools

deped_schools = schools_master[schools_master["school.classification2"] == 'DepED Managed']
lgu_schools = schools_master[schools_master["school.classification2"] == 'Local Government']

deped_schools.isna().sum()
#dropping schools without data on total teachers, rooms, and enrollment
deped_schools.drop(deped_schools[deped_schools['total_teachers'] == 0].index, inplace=True)
deped_schools.drop(deped_schools[deped_schools['school.enrollment'] == 0].index, inplace=True)
deped_schools.drop(deped_schools[deped_schools['total_rooms'] == 0].index, inplace=True)
```

```python
deped_schools['school.region'].value_counts()
```

```python
# visualization for LGU-managed schools per region
spr = deped_schools['school.region'].value_counts()

# indicates if plotting on the figures or on subplots
plt.figure(figsize=(10,6), dpi=200) 

# the main code to create the graph
plt.barh(spr.index, spr.values) 

# additional elements that can be customzed
plt.title("Schools per Region", fontsize = 16)
plt.xlabel("Number of Schools", fontsize=12)
plt.gca().invert_yaxis()

# display graph
plt.show()
```

```python
mps = deped_schools.groupby("school.region")["enrol_mooe"].agg(np.mean)

# indicates if plotting on the figures or on subplots
plt.figure(figsize=(10,6), dpi=200) 

# the main code to create the graph
plt.barh(mps.index, mps.values) 

# additional elements that can be customzed
plt.title("Regional Mean MOOE per Student", fontsize = 16)
plt.xlabel("MOOE per Student (PHP)", fontsize=12)
plt.gca().invert_yaxis()

# display graph
plt.show()
```

```python
mpt = deped_schools.groupby("school.region")["teacher_mooe"].agg(np.mean)

# indicates if plotting on the figures or on subplots
plt.figure(figsize=(10,6), dpi=200) 

# the main code to create the graph
plt.barh(mpt.index, mpt.values) 

# additional elements that can be customzed
plt.title("Regional Mean MOOE per Teacher", fontsize = 16)
plt.xlabel("MOOE per Teacher (PHP)", fontsize=12)
plt.gca().invert_yaxis()

# display graph
plt.show()
```

```python
deped_schools.groupby("school.region")["teacher_mooe"].agg(np.mean)
```

```python
mpr = deped_schools.groupby('school.region')['room_mooe'].sum()

# indicates if plotting on the figures or on subplots
plt.figure(figsize=(10,6), dpi=200) 

# the main code to create the graph
plt.barh(mpr.index, mpr.values) 

# additional elements that can be customzed
plt.title("Mean MOOE per Room", fontsize = 16)
plt.xlabel("MOOE per Room", fontsize=12)
plt.gca().invert_yaxis()

# display graph
plt.show()
```
